```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
import pickle
split_list = [0,1,2,3,4]

dataset = 'adult_race_gender'
save_dir = '../save_fairness/{:s}/'.format(dataset)
seed = 42
shidden ='_512_512_'
str_loss = 'CE'

item = 0


    
for split in split_list:
    savepath = save_dir+'MMPFNN_arq'+shidden+'lossCE_seed'+str(seed)+'_split'+str(split)+'_minimax.pkl'
    with open(savepath, 'rb') as f:
        save_data = pickle.load(f)
    df_table = save_data['group_test_stats'].copy()
    df_table['model']='MMPF NN'
    if item == 0:
        results_table = df_table
    else:
        results_table = pd.concat([results_table,df_table])
    item += 1    

    
for split in split_list:
    savepath = save_dir+'MMPFNN_arq'+shidden+'lossCE_seed'+str(seed)+'_split'+str(split)+'_balanced.pkl'
    with open(savepath, 'rb') as f:
        save_data = pickle.load(f)
    df_table = save_data['group_test_stats'].copy()
    df_table['model']='Balanced'
    if item == 0:
        results_table = df_table
    else:
        results_table = pd.concat([results_table,df_table])
    item += 1   
    
    
for split in split_list:
    savepath = save_dir+'apstar_llr_{:d}.pkl'.format(split)
    with open(savepath, 'rb') as f:
        save_data = pickle.load(f)
    df_table = save_data['group_test_stats'].copy()
    df_table['model']='MMPF LLR'
    if item == 0:
        results_table = df_table
    else:
        results_table = pd.concat([results_table,df_table])
    item += 1   
```
